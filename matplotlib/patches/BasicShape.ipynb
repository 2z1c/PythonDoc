```python
# 
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Ellipse

delta = 45.0  # degrees



# 椭圆心坐标
xy=(1, 1)

# 宽度
width=4
# 高度
height=2
# 角度
angles = np.arange(0, 360 + delta, delta)
ells = [Ellipse(xy=xy, width=width, height=height, angle=a) for a in angles]

a = plt.subplot(111, aspect='equal')

for e in ells:
    e.set_clip_box(a.bbox)
    e.set_alpha(0.1)
    a.add_artist(e)

plt.xlim(-2, 4)
plt.ylim(-1, 3)

plt.show()
```

## 矩形
https://matplotlib.org/api/_as_gen/matplotlib.patches.Rectangle.html#matplotlib.patches.Rectangle

```python
# -*- coding: utf-8 -*-
import matplotlib.pyplot as plt
import matplotlib.patches as patches


# xy = np.random.rand(2)*5
# 左下角坐标
xy = (1,1)
# 宽度
width =1
height=2
fig = plt.figure()
ax = fig.add_subplot(111, aspect='equal')
patche = patches.Rectangle(xy = xy,width = width,height=height)
ax.add_patch(patche)
plt.xlim(-2, 4)
plt.ylim(-1, 3)
# fig.savefig('rect1.png', dpi=90, bbox_inches='tight')
plt.show()
```

## 三角形

```python
import numpy as np 
import matplotlib.pyplot as plt


X = np.array([[1,1], [2,2.5], [3, 1], [8, 7.5], [7, 9], [9, 9]])
Y = ['red', 'red', 'red', 'blue', 'blue', 'blue']

plt.figure()
plt.scatter(X[:, 0], X[:, 1], s = 170, color = Y[:])

t1 = plt.Polygon(X[:3,:], color=Y[0])
plt.gca().add_patch(t1)

t2 = plt.Polygon(X[3:6,:], color=Y[3])
plt.gca().add_patch(t2)

plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.path import Path
codes = []
verts = []
# 下面的步骤表达的是
# 降点移动到 Path.MOVETO 
# 然后根据 Path.LINETO 划线
verts += [(4, 4), (5, 5), (5, 4),(4, 4)]
codes += [Path.MOVETO] + [Path.LINETO]*3


verts += [(3, 3.5), (6, 4), (6, 6),(3, 3.5)]
codes += [Path.MOVETO] + [Path.LINETO]*3

path = Path(verts, codes)
pathpatch = patches.PathPatch(path, facecolor='None', edgecolor='green')
fig, ax = plt.subplots()
ax.add_patch(pathpatch)
ax.set_title("Nothing")

ax.dataLim.update_from_data_xy(verts)
ax.autoscale_view()

# plt.savefig("conpund.png")
plt.show()
```
